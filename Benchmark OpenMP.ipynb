# Benchmark OpenMP

Benchmark to calculate $\pi$ using Leibniz formula with OpenMP 
### Settings
* Number of iterations: $1x10^9$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Time vs Threads

```python
cores = [1, 2, 4, 8, 16, 32]
time = [6.06, 8.08, 9.76, 4.74, 2.3, 1.61]
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.xlabel("Number of threads ($2^n$)")
plt.ylabel("Time (s)")
plt.plot(np.log2(cores), time)
```

## Speedup vs Threads

Speedup is calculated using the equation:
$$S_P(n) = \frac{T^*(n)}{T_P(n)}$$

```python
cores = [1, 2, 4, 8, 16, 32]
speedup = [time[0] / time[i] for i in range(len(time))]
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.xlabel("Number of threads ($2^n$)")
plt.ylabel("Speedup (proportion)")
plt.plot(np.log2(cores), speedup)
```

# Genetic Algorithm Comparison

## Hardware
* 8 Cores
* Intel core i7 7th Gen
* 16 GB RAM

# Experiment 1

## Experiment settings
* **Problem:** Rastrigin Optimization
* **Number of dimensions:** 100 
* **Population size:** 100000
* **Generations:** 50

```python
threads = [1, 2, 4, 8, 16]
time = [43.027, 24.570, 15.097, 13.451, 18.281]
fitness = [190.509, 188.91, 200.486, 197.796, 191.031]
speedup = [time[0] / time[i] for i in range(len(time))]
```

## Time vs Threads

```python
plt.figure(num=None, figsize=(12,8), dpi=80, facecolor='w', edgecolor='k')
plt.xlabel("Number of threads ($2^n$)")
plt.ylabel("Time (s)")
plt.plot(np.log2(threads), time)
```

## Speedup vs Threads

```python
plt.figure(num=None, figsize=(12,8), dpi=80, facecolor='w', edgecolor='k')
plt.xlabel("Number of threads ($2^n$)")
plt.ylabel("Time (s)")
plt.plot(np.log2(threads), speedup)
```

## Summarization

```python
print("|\tThreads\t\t|\tTime\t\t|\tSpeedup\t\t|\tFitness\t\t|")
print("|" + "-"*23 + "|" + "-"*23 + "|" + "-"*23 + "|" + "-"*23 + "|")

for i in range(len(threads)):
    print("|\t{0:10.1f}\t|\t{1:10.3f}\t|\t{2:10.3f}\t|\t{3:10.3f}\t|".format(threads[i], time[i], 
                                                                             speedup[i], fitness[i]))
```

# Benchmark OpenMP

Benchmark to calculate $\pi$ using Leibniz formula with OpenMP 
### Settings
* Number of iterations: $1x10^9$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Time vs Threads

```python
cores = [1, 2, 4, 8, 16, 32]
time = [6.06, 8.08, 9.76, 4.74, 2.3, 1.61]
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.xlabel("Number of threads ($2^n$)")
plt.ylabel("Time (s)")
plt.plot(np.log2(cores), time)
```

## Speedup vs Threads

Speedup is calculated using the equation:
$$S_P(n) = \frac{T^*(n)}{T_P(n)}$$

```python
cores = [1, 2, 4, 8, 16, 32]
speedup = [time[0] / time[i] for i in range(len(time))]
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.xlabel("Number of threads ($2^n$)")
plt.ylabel("Speedup (proportion)")
plt.plot(np.log2(cores), speedup)
```
